```python
import nltk
import random
from nltk.chat.util import Chat, reflections

# NLTK data download
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')

# Define pairs of patterns and responses
pairs = [
    [
        r"my name is (.*)",
        ["Hello %1, how are you today?",]
    ],
    [
        r"what is your name?",
        ["My name is Chatbot and I'm here to assist you.",]
    ],
    [
        r"how are you ?",
        ["I'm doing good\nHow about You ?",]
    ],
    [
        r"sorry (.*)",
        ["It's alright", "No problem",]
    ],
    [
        r"i'm (.*) doing good",
        ["Nice to hear that", "Alright :)",]
    ],
    [
        r"hi|hey|hello",
        ["Hello", "Hey there",]
    ],
    [
        r"(.*) age?",
        ["I'm a computer program, I don't age!",]
    ],
    [
        r"what (.*) want ?",
        ["Make me an offer I can't refuse.",]
    ],
    [
        r"(.*) created ?",
        ["I was created by [Your Name].",]
    ],
    [
        r"(.*) (location|city) ?",
        ["I'm a virtual assistant, I can be wherever you need me to be.",]
    ],
    [
        r"how is the weather in (.*)?",
        ["Weather in %1 is nice.", "It's always sunny in my virtual world!",]
    ],
    [
        r"quit",
        ["Bye, take care. See you soon :) ","It was nice talking to you. See you later!"]
    ],
]

# Create Chatbot
def chatbot():
    print("Hi! I'm Chatbot. How can I assist you today?")
    chat = Chat(pairs, reflections)
    chat.converse()

# Execute chatbot function
if __name__ == "__main__":
    chatbot()
```
